```python
import numpy as np
def replace_plastics_with_noise(image, bounding_boxes):
    # Iterate over the bounding boxes
    for bbox in bounding_boxes:
        bbox = bounding_boxes

        # Extract the region of interest (ROI)
        roi = image[bbox["y"]:bbox["y"]+bbox["h"], bbox["x"]:bbox["x"]+bbox["w"]]

        # Generate random noise
        noise = np.random.randint(0, 256, roi.shape, dtype=np.uint8)

        # Replace the ROI with random noise
        image[bbox["y"]:bbox["y"]+bbox["h"], bbox["x"]:bbox["x"]+bbox["w"]] = noise
```
